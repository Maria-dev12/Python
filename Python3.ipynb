```python
#Siempre lo primero que debo hacer es importar pandas y numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subimos el archivos desde google drive, y se que estoy logeada porque la carpeta esta tachada.
df_estudiantes=pd.read_csv('/content/drive/MyDrive/StudentsPerformance (1).csv')
df_estudiantes
```

```python
#Vamos a crear unas columnas usando el metodo assign o el metodo insert
#Si quiero agregar varias columnas uso assign, y también si quiero sobre escribrir una columna.
filosofia_score=np.random.randint(1,100,size=1000)
fisica_score=np.random.randint(1,100,size=1000)
```

```python
#Ahora agregamos un indice a cada fila para saber en que parte del dataframe
filosofia_nota=pd.Series(filosofia_score, index=np.arange(0,1000))
fisica_nota=pd.Series(fisica_score, index=np.arange(0,1000))
```

```python
#Agregamos las nuevas columnas al dataframe usando assign.
df_estudiantes.assign(filosofia_score=filosofia_nota,fisica_score=fisica_nota)
```

```python
#La tabla anterio es solo un prewiev se debe guardar para sobreescvribir al dataframe
df_estudiantes=df_estudiantes.assign(filosofia_score=filosofia_nota,fisica_score=fisica_nota)
df_estudiantes
```

```python
#Usamos el método insert, el método insert es diferente porqueme permite especificar donde voy a colocar esa columna.
#Al usar el método insert debemos tener cuidado porque no me devuelve un prewiev, si no, el cambio queda realizado de una vez.
df_estudiantes.insert(1,'ED_SCORE',fisica_nota)
```
